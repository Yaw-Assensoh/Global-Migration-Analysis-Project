## Step 1: Import Libraries

```python
# Imorte necessary libraries
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

print(" Fixed imports - ready for professional policy simulation")
```

## Step 2: Clean Data Structure

```python
# create simplified dataset
def create_fixed_data():
    """Fixed data generation without complex structures"""
    countries = ["USA", "Germany", "India", "Nigeria", "Japan", "Brazil"]
    years = list(range(2020, 2025))  # Only historical
    
    data = []
    for country in countries:
        base = {
            "USA": 3.1, "Germany": 1.5, "India": -0.4,
            "Nigeria": -0.2, "Japan": 0.3, "Brazil": -0.2
        }.get(country, 2.0)
        
        for year in years:
            migration = base + (year - 2020) * 0.05 + np.random.normal(0, 0.1)
            data.append({
                'Year': year,
                'Country': country,
                'Migration_Rate': round(migration, 2),
                'Visa_Score': np.random.randint(40, 80),
                'Work_Score': np.random.randint(40, 80)
            })
    
    return pd.DataFrame(data)

df = create_fixed_data()
print(f" Fixed data created: {df.shape}")
print(df.head())
```

## Step 4: Professional Visualization

```python
# Clean, professional visualization that answers the core question

def create_simple_chart(historical_years, historical_rates, 
                       future_years, baseline, scenario,
                       country, visa_change, work_change):
    """Fixed chart without complex annotations"""
    
    fig = go.Figure()
    
    # Historical
    fig.add_trace(go.Scatter(
        x=historical_years, y=historical_rates,
        mode='lines+markers',
        name='Historical',
        line=dict(color='#2E86AB', width=3)
    ))
    
    # Baseline
    fig.add_trace(go.Scatter(
        x=future_years, y=baseline,
        mode='lines',
        name='Baseline',
        line=dict(color='gray', width=3, dash='dash')
    ))
    
    # Scenario
    fig.add_trace(go.Scatter(
        x=future_years, y=scenario,
        mode='lines',
        name=f'Policy Changes (V:{visa_change:+}, W:{work_change:+})',
        line=dict(color='#FF6B6B', width=4)
    ))
    
    # Simple layout
    fig.update_layout(
        title=f'Policy Impact: {country}',
        xaxis_title='Year',
        yaxis_title='Migration Rate',
        height=450,
        plot_bgcolor='white',
        hovermode='x unified'
    )
    
    return fig

print(" Fixed visualization ready")
```
